#### Merge
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html

```python
student = pd.DataFrame({'Name': ['Alan', 'Jack', 'Sparrow', 'new'], 'id': [1,2,3, 5]})
marks = pd.DataFrame({'id': [1,2,3,4], 'Maths': [98,67,86,45], 'Science': [78,96,52,83]})
marks
```

```python
student = pd.merge(student, marks, on = 'id') #default join is 'inner'
student
```

```python
student = pd.DataFrame({'Name': ['Alan', 'Jack', 'Sparrow', 'new'], 'id': [1,2,3, 5]})
pd.merge(student, marks, on = 'id', how = 'right')
```

```python
merged_df = pd.merge(student, marks, on = 'id', how = 'outer')
merged_df
```

### Pivot : Convert data from long to wide format

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.pivot.html

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
ipl = matches[['id', 'season']].merge(deliveries, left_on = 'id', right_on = 'match_id').drop('match_id', axis = 1)
runs_comparison = ipl[ipl.season.isin([2015, 2016])].groupby(['season', 'batsman']).batsman_runs.sum().nlargest(20).reset_index().sort_values(by='batsman')
vc = runs_comparison.batsman.value_counts()
batsmen_comparison = runs_comparison[runs_comparison.batsman.isin(vc[vc == 2].index.tolist())]

batsmen_comparison#.set_index(['batsman', 'season']).unstack()
```

```python
%matplotlib inline
batsmen_comparison.pivot(index='season', columns='batsman', values='batsman_runs').plot(kind = 'bar')
```

```python
batsmen_comparison.pivot(index='batsman', columns='season', values='batsman_runs').plot(kind = 'bar')
```

## Text data: Vectorized string methods
https://pandas.pydata.org/pandas-docs/stable/text.html

```python
df = pd.DataFrame({'Team': ['100rcb5', 'dd2%', 'kkr@3', 'KKR3', 'RCB1', 'SRH4', 'MI5','KXIP6', 'rps7', 'dd2', 'DD2', 'mi5'],
                  'Player': ['virat_kohli', 'zaheer_khan', 'Gautam_Gambhir', 'sunil_narine','chris_Gayle', 'david_warner', 'rohit_sharma', 'yusuf_pathan', 'ms_dhoni', 'jp_duminy', 'amit_mishra', 'harbhajan_singh']})
df
```

#### str.upper

```python
df.Team = df.Team.str.upper() #assignment is needed
df#.Team.value_counts()
```

#### str.extract

```python
df.Team =  df.Team.str.extract('([A-Z]+)', expand = False) #\d+ or [0-9]
df.Team.value_counts()
```

#### str.replace

```python
#replace
df.Player = df.Player.str.replace('_', ' ').str.title() #chaining the operations
df
```

#### str.split

```python
df['First Name'] = df.Player.str.split().str[0] #default is expand = False
df['Last Name'] = df.Player.str.split().str[1] #default is expand = False
df.drop('Player', axis = 1, inplace = True)
df
```

### Extract Using Regex (Regular Expression)

    * 0 or more
    + 1 or more
    .* : 0 or more
    .+ : 1 or more
   
    
    \s+ : 1 or more space
    \s{3} space
    
    [A-Za-z] : all alphabets
    [A-Za-z ] : all alphabets including the space
    
    \d+ : all numbers
    
    [A-Za-z0-9] : alpha-numeric
    \w+ : alpha-numeric
    
    \W+ : everything except alpha-numeric

```python
#Extract all the values after the underscore
df = pd.DataFrame({'Team': ['100rcb5', 'dd2%', '^4kkr@3', 'KKR3', 'RCB1', 'SRH4', 'MI5','KXIP6', 'rps7', 'dd2', 'DD2', 'mi5'],
                  'Player': ['virat_kohli', 'zaheer_khan', 'Gautam_Gambhir', 'sunil_narine','chris_Gayle', 'david_warner', 'rohit_sharma', 'yusuf_pathan', 'ms_dhoni', 'jp_duminy', 'amit_mishra', 'harbhajan_singh']})

df[['first', 'last']]=df.Player.str.extract('(.*)_(.*)', expand = True)
df.head()
```

```python
#Extract Team code number from Team column
df.Team.str.extract('(\d+)', expand = True).head() #\d+ to extract all numeric values

df.Team.str.replace('[A-Za-z|\W]', '')
```

```python
df.Player.str.extract('([A-Za-z ]+)', expand = True).head() #\d+ to extract all alphabets
```

```python
df.Team.str.extract('(\w+)', expand = True).head() #\w+ to extract all alpha-numeric values
```

```python
df.Team.str.extract('(\W+)', expand = True).head() #\d+ to extract all non-alpha-numeric values
```

#### Extractall: Extract all matches in each subject
##### The extractall method returns every match. The result of extractall is always a DataFrame with a MultiIndex on its rows. The last level of the MultiIndex is named match and indicates the order in the subject.

```python
df.Team.str.extractall('(\d+)?([A-Za-z]+)(\d+)')
```

```python
df[['First_Name', 'Last_Name']] = df.Player.str.extractall('([A-Za-z]+)_([A-Za-z]+)').reset_index(level=-1, drop=True)
df
```
